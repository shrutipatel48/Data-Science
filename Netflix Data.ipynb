# Monthly and Yearly Releasing TV shows and Movies

```python
df['year']=df['date_added'].dt.year
df['month']=df['date_added'].dt.month
df['day']=df['date_added'].dt.day
```

```python
plt.figure(figsize=(10,8))
monthly_movie=df[df['type']=='Movie']['month'].value_counts().sort_index()
monthly_series=df[df['type']=='TV Show']['month'].value_counts().sort_index()

plt.plot(monthly_movie.index, monthly_movie.values, label='Movies',linewidth=3,marker='*',markersize=10)
plt.plot(monthly_series.index, monthly_series.values,label='Series',linewidth=3,marker='*',markersize=10)

plt.xlabel("Months")
plt.ylabel("Frequency of releases")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.suptitle("Monthly releases of Movies and TV shows on Netflix",size=25,style = 'italic')
plt.show()
```

```python
plt.figure(figsize=(10,8))
yearly_movie=df[df['type']=='Movie']['year'].value_counts().sort_index()
yearly_series=df[df['type']=='TV Show']['year'].value_counts().sort_index()

plt.plot(yearly_movie.index, yearly_movie.values, label='Movies',linewidth=3,marker='.',markersize=15)
plt.plot(yearly_series.index, yearly_series.values,label='Series',linewidth=3,marker='.',markersize=15)

plt.xlabel("Years")
plt.ylabel("Frequency of releases")
plt.legend()
plt.suptitle("Yearly releases of Movies and TV shows on Netflix",size=25,style = 'italic')
plt.show()
```
